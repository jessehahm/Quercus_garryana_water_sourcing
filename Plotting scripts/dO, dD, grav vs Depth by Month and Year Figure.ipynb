```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
from matplotlib import cm
import matplotlib.cm as cmx
import matplotlib.colors as colors
plt.rcParams["font.sans-serif"] = "Arial"
plt.rcParams['pdf.fonttype'] = 42
plt.rcParams['ps.fonttype'] = 42
import datetime as dt
plt.rcParams['figure.autolayout'] = True

plt.rcParams['mathtext.fontset'] = 'stix'
plt.rcParams['mathtext.rm'] = 'Arial'
plt.rcParams['mathtext.it'] = 'Arial'
plt.rcParams['mathtext.bf'] = 'Helvetica:bold'
plt.rcParams['font.size'] = 8
import matplotlib as mpl

%matplotlib inline
```

```python
df = pd.read_csv('../Data/isotope-data.csv',
                 sep=',',  skiprows = [1])
df['Date Collected'] = pd.to_datetime(df['Date Collected'])
```

```python
sagehorn = df[df['Site']=='Sagehorn']

drilled = sagehorn[(sagehorn['Sample Type II']=='Drilled material')]

drilled['day'] = drilled['Date Collected'].dt.date


drilled['dayofyear'] = drilled['Date Collected'].dt.dayofyear

#drilled.Location.fillna(drilled['Sample Type II'], inplace=True)
drilled['edgecolors']='grey'
drilled.loc[drilled['Energetic State'] == 'saturated', 'edgecolors'] = 'gold'
drilled['Depth (cm)']=drilled['Depth (cm)'].astype(float)
```

```python
gw = sagehorn[sagehorn['Sample Type II']=='Well water']
stems = sagehorn[sagehorn['Sample Type II']=='Stem']
stems['day'] = stems['Date Collected'].dt.date
stems['dayofyear'] = stems['Date Collected'].dt.dayofyear
# stems = stems[(stems['Date Collected'] > pd.to_datetime('2015-01-01 00:00:00'))]

gw = gw[(gw['Location'] == 'MN1') | 
                      (gw['Location'] == 'MN-1') |
                      (gw['Location'] == 'MS4') |
                      (gw['Location'] == 'MS-4')|
                      (gw['Location'] == 'W507')]
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, sharey=True, sharex=False, figsize=(8,4))
fig.set_tight_layout(False)


days = sorted((set(drilled['day'])))

def truncate_colormap(cmap, minval=0.0, maxval=1.0, n=100):
    new_cmap = colors.LinearSegmentedColormap.from_list(
        'trunc({n},{a:.2f},{b:.2f})'.format(n=cmap.name, a=minval, b=maxval),
        cmap(np.linspace(minval, maxval, n)))
    return new_cmap

cmap = plt.cm.get_cmap('nipy_spectral')
cmap = truncate_colormap(cmap, 1.0/8, 1)
cmap = colors.LinearSegmentedColormap.from_list('mymap',
                                    ['midnightblue','#4393c3','slategrey','#fddbc7',
                                     'orange','#b2182b','white','slateblue'])
bounds = [92,122,153, 183, 214, 245, 275, 306, 336]
norm = colors.BoundaryNorm(bounds, cmap.N)

ax1.scatter(1000,1000,marker = 's',c='white',edgecolors='black',s=30,label='2015')
ax1.scatter(1000,1000,marker = 'o',c='white',edgecolors='black',s=50,label='2016')
ax1.scatter(1000,1000,marker = '^',c='white',edgecolors='black',s=45,label='2017')
ax1.scatter(1000,1000,marker = 'D',c='white',edgecolors='black',s=30,label='2018')

ax1.legend(loc='lower right',scatterpoints=1)

ax1.axvspan(gw.dO.mean()-gw.dO.std(), gw.dO.mean()+gw.dO.std(), facecolor='blue', alpha=0.3,zorder=-10,label='Groundwater mean+/-1 std.')
ax2.axvspan(gw.dD.mean()-gw.dD.std(), gw.dD.mean()+gw.dD.std(), facecolor='blue', alpha=0.3,zorder=-10,label='Groundwater')

ax1.axhspan(60,35, facecolor='grey', alpha=0.2,zorder=-10)
ax2.axhspan(60,35, facecolor='grey', alpha=0.2,zorder=-10)
ax3.axhspan(60,35, facecolor='grey', alpha=0.2,zorder=-10)


#PLOT SUBSURFACE MATERIALS
basesize = 20
for day in days:

    thisDay = drilled[drilled['day']==day]
    thisDay = thisDay.sort_values(['Depth (cm)'], na_position='last')
    if day.year == 2016:
        marker = 'o'
        size=basesize+20
    if day.year == 2015:
        marker = 's'
        size=basesize
    if day.year == 2017:
        marker = '^'
        size=basesize+15
    if day.year == 2018:
        marker = 'D'
        size=basesize        
    plotData = thisDay[pd.isnull(thisDay.dO)==False] 
    ax1.scatter(plotData.dO, plotData['Depth (cm)'],
                         marker=marker,
                         c=plotData['dayofyear'], #alpha=0.8,
                         norm=norm,
                         cmap=cmap,
                         edgecolors=plotData['edgecolors'],
                         zorder = 100, s = size,clip_on=True)
    ax2.scatter(thisDay.dD, thisDay['Depth (cm)'],
                         marker=marker,
                         c=thisDay['dayofyear'], #alpha=0.8,
                         norm=norm,
                         cmap=cmap,
                         edgecolors=thisDay['edgecolors'],
                         zorder = 100, s = size,clip_on=True)
    plotData = thisDay[pd.isnull(thisDay['Gravimetric moisture content (g/g)'])==False] 
    ax3.scatter(plotData['Gravimetric moisture content (g/g)'], plotData['Depth (cm)'],
                         marker=marker,
                         c=plotData['dayofyear'], #alpha=0.8,
                         norm=norm,
                         cmap=cmap,
                         edgecolors=plotData['edgecolors'],
                         zorder = 100, s = size,clip_on=True)

ax1.text(0.03, 0.85, '(a)',
        verticalalignment='top', horizontalalignment='left',
        transform=ax1.transAxes,
        color='black', fontsize=10)

# ax1.text(0, 1.4, 'Q. garryana stems',        rotation='vertical',
#         verticalalignment='top', horizontalalignment='left', style='italic',
#         transform=ax1.transAxes,
#         color='black', fontsize=10)


ax2.text(0.03, 0.85, '(b)',
        verticalalignment='top', horizontalalignment='left',
        transform=ax2.transAxes,
        color='black', fontsize=10)
ax3.text(0.03, 0.85, '(c)',
        verticalalignment='top', horizontalalignment='left',
        transform=ax3.transAxes,
        color='black', fontsize=10)

ax3.text(0.3, 0.25, 'symbol edge color:',style='italic',
        verticalalignment='top', horizontalalignment='left',
        transform=ax3.transAxes,bbox={'facecolor':'white','edgecolor':'white', 'alpha':1, 'pad':.1},
        color='black', fontsize=10)

ax3.text(0.5, 0.2, 'unsaturated',
        verticalalignment='top', horizontalalignment='left',
        transform=ax3.transAxes,bbox={'facecolor':'white','edgecolor':'white', 'alpha':1, 'pad':.1},
        color='grey', fontsize=10)
ax3.text(0.5, 0.15, 'saturated',
        verticalalignment='top', horizontalalignment='left',
        transform=ax3.transAxes,bbox={'facecolor':'white','edgecolor':'white', 'alpha':1, 'pad':.1},
        color='gold', fontsize=10)

ax=plt.gca() #get the current axes
PCM=ax.get_children()[2] #get the mappable, the 1st and the 2nd are the x and y axes

cb =plt.colorbar(PCM, ax=ax,fraction=0.046,# pad=0.04,
                                cmap=cmap, norm=norm, 
                                ticks=np.array(bounds[0:-1])+15,
                                boundaries=bounds)
cb.ax.set_yticklabels(['April','May','June','July','August','September','October (no data)','November'])  # vertically oriented colorbar
cb.ax.invert_yaxis() 

ax1.grid(alpha=0.3)
ax1.set_ylabel('Depth (cm)')
ax1.set_xlabel(r'$\delta^{18}$O' + u'\u2030')
ax1.set_xlim([-15,2])
 
          
        
ax2.grid(alpha=0.3)
ax2.set_xlabel(r'$\delta$D' + u'\u2030')
ax2.set_xlim([-80,-20])

ax3.grid(alpha=0.3)
ax3.set_xlabel('Cryogenic-extraction inferred \n gravimetric moisture content (g/g)')
ax3.set_xlim([0,.4])
ax3.set_xticks([0,0.1,0.2,0.3,.4])   

ax1.set_ylim(0,300)
ax1.set_yticks(np.arange(0,300,50))
ax1.set_yticklabels([x for x in ax1.get_yticks()])

ax1.invert_yaxis() 

savedFile = '../Figs/DrilledMaterial_depthPlots.pdf'


plt.savefig(savedFile, bbox_inches='tight')
```
