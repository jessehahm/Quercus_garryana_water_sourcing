```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
from matplotlib import cm
import matplotlib.cm as cmx
import matplotlib.colors as colors
plt.rcParams["font.sans-serif"] = "Arial"
plt.rcParams['pdf.fonttype'] = 42
plt.rcParams['ps.fonttype'] = 42
import datetime as dt
plt.rcParams['figure.autolayout'] = True

plt.rcParams['mathtext.fontset'] = 'stix'
plt.rcParams['mathtext.rm'] = 'Arial'
plt.rcParams['mathtext.it'] = 'Arial'
plt.rcParams['mathtext.bf'] = 'Helvetica:bold'
plt.rcParams['font.size'] = 8
import matplotlib as mpl

%matplotlib inline
```

```python
df = pd.read_csv('../Data/isotope-data.csv',
                 sep=',',  skiprows = [1])
df['Date Collected'] = pd.to_datetime(df['Date Collected'])
```

```python
sagehorn = df[df['Site']=='Sagehorn']

sagehorn['day'] = sagehorn['Date Collected'].dt.date
```

```python
QUGA = sagehorn[sagehorn['Species (scientific)']=='QUGA']
stems = QUGA[QUGA['Sample Type II']=='Stem']
cores = QUGA[QUGA['Sample Type II']=='Tree core']

```

```python

joined = stems.merge(cores,how='inner',on=['Date Collected','Location'],suffixes=('_stems','_cores'))
joined=joined[joined['Date Collected']>pd.to_datetime('2016-01-01')]
```

```python
joined['date'] = joined['Date Collected'].dt.date
joined
```

```python
fig, (ax) = plt.subplots(1, 1,  figsize=(4,4))

dates = sorted(set(joined['date']))
markers = ['^', 'o', 's']
s=50
for i, date in enumerate(dates):
    plotData = joined[joined['date']==date]
    plt.plot([plotData['dO_cores'].astype('float'), plotData['dO_stems'].astype('float')],
             [plotData['dD_cores'].astype('float'), plotData['dD_stems'].astype('float')],c='black',zorder=5)
    
    ax.scatter(plotData['dO_cores'].astype('float'),plotData['dD_cores'].astype('float'),
               label='Cores, '+ str(date),color='white', marker=markers[i],edgecolor='black',s=s,zorder=10)
    ax.scatter(plotData['dO_stems'].astype('float'),plotData['dD_stems'].astype('float'), 
               label='Stems, '+ str(date),color='black', marker=markers[i],edgecolor='black',s=s,zorder=10)

plt.legend(scatterpoints=1,loc='upper left', fontsize=8)

ax.set_ylabel(r'$\delta$D' + u'\u2030')
ax.set_xlabel(r'$\delta^{18}$O' + u'\u2030')

# plt.title('Quercus garryana tissue comparison')
plt.grid(alpha=.3)
savedFile = '../Figs/SagehornIsotopes_QUGA_Stems-Vs-Cores.pdf'
plt.savefig(savedFile, bbox_inches='tight')
# os.startfile(savedFile)
plt.show()
```
